```python
import grid2op
import l2rpn_baselines
from l2rpn_baselines import DoNothing, DeepQSimple, DoubleDuelingDQN, DoubleDuelingRDQN, DuelQLeapNet, DuelQSimple, Geirina, PandapowerOPFAgent, SAC, SliceRDQN, AsynchronousActorCritic
```

```python
envs = ['rte_case14_realistic', 'l2rpn_neurips_2020_track2_small']
models = [DoNothing, DeepQSimple, DoubleDuelingDQN, DoubleDuelingRDQN, 
          DuelQLeapNet, DuelQSimple, Geirina, PandapowerOPFAgent, SAC, SliceRDQN, AsynchronousActorCritic]

env_id = 0
model_id = 2

train_iters = 100
eval_nb_episodes = 10
```

### 1. Settings

```python
print(f'Environment = {envs[env_id]}\n\nModel = {models[model_id].__name__}')
```

```python
env = grid2op.make(envs[env_id])
model = models[model_id]
train = model.train
evaluate = model.evaluate
```

### 2. Train

```python
train(env, iterations=100)
```

### 3. Evaluate

```python
evaluate(env, load_path='models/DoubleDuelingDQN.h5', nb_episode=eval_nb_episodes)
```
